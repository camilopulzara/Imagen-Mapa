```python
import folium
import pandas as pd
import numpy as np
from folium import plugins
from folium.plugins import HeatMap
```

```python
data = pd.read_excel('C:/Users/pulzara/Documents/bancolombia/entrenamiento_precios_vivienda.xlsx', engine='openpyxl')   
```

```python
data1 = pd.read_table('C:/Users/pulzara/Documents/bancolombia/PuntosInteres.csv', encoding='ISO-8859-1', sep=';', header=None)            
```

```python
data1.rename(columns={0:'Longitud', 1:'Latitud', 2:'Punto_interes', 3:'Nombre'}, inplace=True )
```

```python
data1 = data1.dropna() 
```

```python
data1.reset_index(drop = True, inplace=True)
```

```python
df = data[['Longitud', 'Latitud', 'tipo_inmueble', 'valor_total_avaluo']]
```

```python
df = df.dropna() 
```

```python
df = df.tail(200) 
data2 = data1.head(200)
```

```python
world_map = folium.Map(location =[4.570868, -74.297333], tiles='stamenterrain', zoom_start=10,prefer_canvas=True)
coordenadas = []

for i in range(0, 200): #representaremos los 100 primeros 
    

    folium.Marker(location=[df.iloc[i]['Latitud'], df.iloc[i]['Longitud']],
                popup=('Tipo de propiedad: '+ df.iloc[i]['tipo_inmueble'] + '\nPrecio: ' + str(df.iloc[i]['valor_total_avaluo'])),     
                icon=folium.Icon(icon='home', color='blue')
                 ).add_to(world_map)
     
    folium.CircleMarker(location=[data2.iloc[i]['Latitud'], data2.iloc[i]['Longitud']],
                        popup = folium.Popup(('Punto de interes: '+ data2.iloc[i]['Punto_interes'] + '\nNombre: ' + str(data2.iloc[i]['Nombre'])), parse_html=True),        
                radius = 15, color = 'red',fill= True,  #icon=folium.Icon(icon='info-sign', color='red')
                  ).add_to(world_map) 
   
    
    coordenadas.append([df.iloc[i]['Latitud'], df.iloc[i]['Longitud']]) 
    

#world_map.add_child(folium.ClickForMarker(popup= "Awesome"))
baseMap = world_map.add_child(plugins.HeatMap(coordenadas)) 
 
baseMap
```

***En el mapa se puede observar las propiedades junto a los puntos de interes, los cuales la persona este interesada en vivir cerca de su trabajo,  de un hospital, de una farmacia, entre otras, lo podra observar de manera sencilla y visual para tomar mejores decisiones a la hora de comprar un inmueble.***
